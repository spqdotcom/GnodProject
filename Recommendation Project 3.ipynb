```python
def create_summary_excel(df):
    column_summary_df = pd.DataFrame({
        'Column Name': df.columns,
        'Data Type': df.dtypes.values,
        'Percentage Null': df.isnull().mean().values * 100,
        'Unique Values': df.nunique().values,
        'Min': np.nan,
        'Max': np.nan,
        'Mean': np.nan,
        'Median': np.nan,
        'Std': np.nan,
        'Unique Content': ''
    })

    include_stats = input("Do you want to include statistical data and unique values? (yes/no): ").strip().lower()

    if include_stats == 'yes':
        for i, col in enumerate(df.columns):
            if pd.api.types.is_numeric_dtype(df[col]):
                column_summary_df.at[i, 'Min'] = df[col].min()
                column_summary_df.at[i, 'Max'] = df[col].max()
                column_summary_df.at[i, 'Mean'] = df[col].mean()
                column_summary_df.at[i, 'Median'] = df[col].median()
                column_summary_df.at[i, 'Std'] = df[col].std()
                # Add unique values if they are less than or equal to 50
                if df[col].nunique() <= 50:
                    column_summary_df.at[i, 'Unique Content'] = df[col].unique()

            elif pd.api.types.is_object_dtype(df[col]):
                # Add all unique values for object data type columns
                column_summary_df.at[i, 'Unique Content'] = df[col].unique()

    save_to_excel = input("Do you want to save the summary to an Excel file? (yes/no): ").strip().lower()
    
    if save_to_excel == 'yes':
        file_name = input("Enter the file name (without extension): ").strip()
        full_file_name = f"{file_name}.xlsx"
        column_summary_df.to_excel(full_file_name, index=False, engine='openpyxl')
        print(f"File saved as {full_file_name}")
    else:
        print("Summary not saved to an Excel file.")
    
    print("\nSummary DataFrame:")
    print(column_summary_df)
```

```python
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials
import time
import re

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=config.client_id,
                                                         client_secret=config.client_secret))

def clean_artist(artist):
    # Remove "Featuring" and everything after it for the initial search
    return re.split(r'\s+Featuring\s+', artist, 1)[0].strip()

spotify_data_retries = []

for index, row in df_nulls_spotify_id.iterrows():
    try:
        # Clean the artist name
        primary_artist = clean_artist(row['artist'])
        query = f"track:{row['song']} artist:{primary_artist}"
        
        results = sp.search(q=query, type='track', limit=5, market="US")
        
        if results['tracks']['items']:
            best_match = None
            # Look for the best match by checking artist names
            for track in results['tracks']['items']:
                track_artists = [artist['name'].lower() for artist in track['artists']]
                if primary_artist.lower() in track_artists:
                    # Check if featured artists are in the original string (case-insensitive)
                    full_artist_lower = row['artist'].lower()
                    if all(any(feat.lower() in full_artist_lower for feat in track_artists) 
                           for feat in track_artists if feat != primary_artist.lower()) or len(track['artists']) == 1:
                        best_match = track
                        break
            else:
                best_match = next((track for track in results['tracks']['items'] 
                                 if primary_artist.lower() in [a['name'].lower() for a in track['artists']]), 
                                 results['tracks']['items'][0])

            track_info = {
                'song': row['song'],
                'artist': row['artist'],
                'spotify_id': best_match['id'],
                'popularity': best_match['popularity'],
                'duration_ms': best_match['duration_ms'],
                'album': best_match['album']['name'],
                'release_date': best_match['album']['release_date'],
                'spotify_uri': best_match['uri'],
                'matched_artists': ", ".join([artist['name'] for artist in best_match['artists']])
            }
            spotify_data_retries.append(track_info)
            print(f"Found match for {row['song']} by {row['artist']}: {track_info['spotify_id']}")
        else:
            print(f"No results found for {row['song']} by {row['artist']}")
            spotify_data_retries.append({
                'song': row['song'],
                'artist': row['artist'],
                'spotify_id': None,
                'popularity': None,
                'duration_ms': None,
                'album': None,
                'release_date': None,
                'spotify_uri': None,
                'matched_artists': None
            })
        
        time.sleep(0.2)  # Slightly longer delay for retries
        
    except Exception as e:
        print(f"Error processing {row['song']} by {row['artist']}: {str(e)}")
        spotify_data_retries.append({
            'song': row['song'],
            'artist': row['artist'],
            'spotify_id': None,
            'popularity': None,
            'duration_ms': None,
            'album': None,
            'release_date': None,
            'spotify_uri': None,
            'matched_artists': None
        })

retried_spotify_df = pd.DataFrame(spotify_data_retries)

expected_columns = ['song', 'artist', 'spotify_id', 'popularity', 'duration_ms', 
                    'album', 'release_date', 'spotify_uri', 'matched_artists']
retried_spotify_df = retried_spotify_df[expected_columns]

print("\nResults for the retried songs:")
print(retried_spotify_df)
```

```python
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyClientCredentials
import time
import re

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=config.client_id,
                                                         client_secret=config.client_secret))

data = [
    {"song": "I Know Love", "artist": "Tate McRae Featuring The Kid LAROI"},
    {"song": "4x4xU", "artist": "Lainey Wilson"}
]
df_failed = pd.DataFrame(data)

def clean_artist(artist):
    if "Featuring" in artist:
        return re.split(r'\s+Featuring\s+', artist, 1)[0].strip()
    return artist.strip()

spotify_data = []

for index, row in df_failed.iterrows():
    try:
        primary_artist = clean_artist(row['artist'])
        query = f"track:{row['song']} artist:{primary_artist}"
        print(f"Trying query: {query}")
        
        results = sp.search(q=query, type='track', limit=5, market="US")
        print(f"Found {results['tracks']['total']} total results")
        
        if results['tracks']['items']:
            best_match = None
            for track in results['tracks']['items']:
                track_artists = [artist['name'].lower() for artist in track['artists']]
                # For "I Know Love", ensure both artists are present if "Featuring" is in the original
                if "featuring" in row['artist'].lower():
                    if (primary_artist.lower() in track_artists and 
                        "the kid laroi" in track_artists):
                        best_match = track
                        break
                elif primary_artist.lower() in track_artists:
                    best_match = track
                    break
            
            if not best_match and "Featuring" in row['artist']:
                query = f"track:{row['song']} {row['artist'].replace('Featuring', '')}"
                print(f"Trying fallback query: {query}")
                results = sp.search(q=query, type='track', limit=1, market="US")
                if results['tracks']['items']:
                    best_match = results['tracks']['items'][0]
            
            if best_match:
                track_info = {
                    'song': row['song'],
                    'artist': row['artist'],
                    'spotify_id': best_match['id'],
                    'popularity': best_match['popularity'],
                    'duration_ms': best_match['duration_ms'],
                    'album': best_match['album']['name'],
                    'release_date': best_match['album']['release_date'],
                    'spotify_uri': best_match['uri'],
                    'matched_artists': ", ".join([artist['name'] for artist in best_match['artists']])
                }
                spotify_data.append(track_info)
                print(f"Success: {row['song']} - Spotify ID: {track_info['spotify_id']}")
            else:
                print(f"No suitable match found for {row['song']} by {row['artist']}")
                spotify_data.append({
                    'song': row['song'],
                    'artist': row['artist'],
                    'spotify_id': None,
                    'popularity': None,
                    'duration_ms': None,
                    'album': None,
                    'release_date': None,
                    'spotify_uri': None,
                    'matched_artists': None
                })
        else:
            print(f"No results for {row['song']} by {row['artist']}")
            spotify_data.append({
                'song': row['song'],
                'artist': row['artist'],
                'spotify_id': None,
                'popularity': None,
                'duration_ms': None,
                'album': None,
                'release_date': None,
                'spotify_uri': None,
                'matched_artists': None
            })
        
        time.sleep(0.2)
        
    except Exception as e:
        print(f"Error: {row['song']} by {row['artist']} - {str(e)}")
        spotify_data.append({
            'song': row['song'],
            'artist': row['artist'],
            'spotify_id': None,
            'popularity': None,
            'duration_ms': None,
            'album': None,
            'release_date': None,
            'spotify_uri': None,
            'matched_artists': None
        })

retried_df = pd.DataFrame(spotify_data)
expected_columns = ['song', 'artist', 'spotify_id', 'popularity', 'duration_ms', 
                    'album', 'release_date', 'spotify_uri', 'matched_artists']
retried_df = retried_df[expected_columns]

print("\nFinal Results:")
print(retried_df)
```
